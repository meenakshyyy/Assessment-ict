## Diabetes Dataset EDA:

* Univariate Analysis:
* Analyze individual variables (e.g., histograms, box plots).
* Bivariate Analysis:
* Explore relationships between pairs of variables (e.g., scatter plots, correlation matrices).
* Multivariate Analysis:
* Investigate interactions among multiple variables (e.g., pair plots, 3D plots).

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("diabetes_model.csv")
```

## Univariate Analysis

```python
col=["Pregnancies",	"Glucose",	"BloodPressure",	"SkinThickness",	"Insulin",	"BMI",	"DiabetesPedigreeFunction"]
df[col].hist(bins=25,figsize=(10,10))
```
